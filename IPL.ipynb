```python
# Plot to visualise the no. of matches held every season
sns.countplot(ipl['year'])
```

```python
# Plot to visualise the no. of matches won by each team
data = ipl.winner.value_counts()
sns.barplot(y = data.index, x = data, orient='h')
```

```python
# Plot to visualise the top 10 players based on player of the match
top_players = ipl.player_of_match.value_counts()[:10]
fig, ax = plt.subplots()
ax.set_xlim([0,20])
ax.set_xlabel("Count")
ax.set_title("Top player of the match Winners")
sns.barplot(y = top_players.index, x = top_players, orient='h')
plt.show()
```

```python
# Plot to visualise the no. of matches held in each city
sns.countplot(y='city', data=ipl)
plt.title('No. of matches held in each city\n')
plt.xlabel('\nNo. of matches held')
plt.ylabel('Cities\n')
plt.xlim([0,120])
plt.show()
```

```python
# Plot to visualise the no. of tosses won by each team
toss=ipl['toss_winner'].value_counts()
fig, ax=plt.subplots()
ax.set_title('No. of tosses won by each team\n')
sns.barplot(y=toss.index, x=toss, orient='h')
plt.xlabel('\nNo. of tosses won')
plt.ylabel('Teams\m')
plt.show()
```

```python
# No. of matches played in different stadiums
sns.countplot(y='venue', data=ipl)
plt.title('No. of matches played in different stadiums\n')
plt.xlabel('\nNo. of matches played')
plt.ylabel('Stadiums\n')
plt.show()
```

```python
matches_per_season = ipl.groupby('year').id.count()
```

```python
toss_decision_percentage = ipl.groupby('year').toss_decision.value_counts().sort_index() / matches_per_season * 100
```

```python
toss_decision_percentage.unstack().plot(kind = 'bar', figsize=(12,6), title = 'Toss Decisions', xlabel = 'Seasons', ylabel = 'Percentage')
```

```python
matches_won_each_season = pd.crosstab(ipl['winner'], ipl['year'])
```

```python
plt.figure(figsize=(15, 9))
plt.xlabel('Seasons')
plt.ylabel('Teams')
plt.title('Matches Won Each Season')
sns.heatmap(matches_won_each_season, annot = True, cmap = 'flare', fmt = 'd', cbar_kws={"orientation": "horizontal"})
```
